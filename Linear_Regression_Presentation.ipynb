```python
import numpy as np
import matplotlib.pyplot as plt
from module.Linear_Regression import LinearRegression
```

# Data Creating.

```python
# Generating 200 random points from uniform distribution.
X=np.random.uniform(0,1,200).reshape(200,1)
np.random.seed(54)
# Initializing b_0 and b_1.
b_0 = 3
b_1 = 7
noise = np.random.normal(0,0.4,200).reshape(200,1)
# Calculating Y with b_0, b_1 and randomly generated noise.
Y=b_0+b_1*X+noise
# Plotting the data created.
plt.scatter(X,Y)
plt.show()
```

# Linear Regression: Least Squares method.

```python
LR = LinearRegression()
LR.fit_least_squares(X, Y)
Y_pred = LR.predict(X)
print("Loss: " + str(LR.loss(Y)))
LR.plot(X, Y, Y_pred)
```

# Linear Regression: SVD method (Singular-Value Decomposition).

```python
LR = LinearRegression()
LR.fit_svd(X, Y)
Y_pred = LR.predict(X)
print("Loss: " + str(LR.loss(Y)))
LR.plot(X, Y, Y_pred)
```

# Linear Regression: Gradient Descent method (Batch Gradient Descent).

```python
LR = LinearRegression()
LR.fit_gradient_descent(X, Y)
Y_pred = LR.predict(X)
print("Loss: " + str(LR.loss(Y)))
LR.plot(X, Y, Y_pred)
```

# Linear Regression: Maximum Likelihood method.

### The solution for this method is the same as the solution for the Least Squares method, just because maximizing the likelihood results the same as minimizing the MSE (Mean Squared Error).

```python
LR = LinearRegression()
LR.fit_maximum_likelihood(X, Y)
Y_pred = LR.predict(X)
print("Loss: " + str(LR.loss(Y)))
LR.plot(X, Y, Y_pred)
```
